```python
import win32com.client
sh=win32com.client.gencache.EnsureDispatch('Shell.Application',0)
ns = sh.NameSpace(r'C:\Users\arvin\Desktop\project\intelligent survillence system\project\video-surviellance-project-code\train\avenue_dataset\training_videos')
colnum = 0
columns = []
while True:
    colname=ns.GetDetailsOf(None, colnum)
    if not colname:
        break
    columns.append(colname)
    colnum += 1

list1=list()
for item in ns.Items():
    print (item.Path)
    for colnum in range(len(columns)):
        colval=ns.GetDetailsOf(item, colnum)
        if colval:
            list1.append(colval)
            print('\t', columns[colnum], colval)
        else:
            list1.append('NA')
            print('\t', columns[colnum], 'NA')
    print("\n")
    print("\n")
    
    
f_name=[a for a in list1[0::37]]
size=[a for a in list1[1::37]]
Item_type=[a for a in list1[2::37]]
Date_created=[a for a in list1[3::37]]
Date_modified=[a for a in list1[4::37]]
perceived_type=[a for a in list1[9::37]]
kind=[a for a in list1[11::37]]
length=[a for a in list1[27::37]]

import csv
with open('train_file_properties.csv', mode='w') as csv_file:
    fieldnames = ['Name', 'Size', 'Item type', 'Date modified', 'Date created', 'perceived_type', 'kind','length']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for i in range(16):
        writer.writerow({"Name":f_name[i], "Size":size[i], 'Item type':Item_type[i], 'Date modified':Date_modified[i],
                         'Date created':Date_created[i],'perceived_type':perceived_type[i],'kind':kind[i],'length':length[i]})
```

```python
import pandas as pd
train_df=pd.read_csv("train_file_properties.csv")
train_df=pd.DataFrame(data=train_df)
print("Details of all files present in training folder")
train_df.head(10)
```
